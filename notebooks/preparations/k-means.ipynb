```python
import pandas as pd
import sys
import os
import pandas as pd
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, adjusted_rand_score
from scipy.optimize import linear_sum_assignment
from sklearn.preprocessing import StandardScaler

os.environ["LOKY_MAX_CPU_COUNT"] = "8"  # Replace 8 with your number of logical cores


sys.path.append(os.path.abspath('../'))
```

```python
file_path = 'data/preprocessed/1480/data_processed.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
clusters_file_path = 'data/preprocessed/1480/clusters.csv'

y_true = pd.read_csv(clusters_file_path)

y_true.head()
```

```python
from sklearn.metrics import confusion_matrix

# KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
y_pred = kmeans.fit_predict(df)

# Use Hungarian method to find best label permutation
def cluster_accuracy(y_true, y_pred):
    # Create contingency matrix
    contingency_matrix = confusion_matrix(y_true, y_pred)
    
    # Apply Hungarian algorithm
    row_ind, col_ind = linear_sum_assignment(-contingency_matrix)  # maximize the match
    
    # Compute accuracy
    optimal_mapping = dict(zip(col_ind, row_ind))
    y_aligned = [optimal_mapping[label] for label in y_pred]
    return accuracy_score(y_true, y_aligned), y_aligned

accuracy, y_aligned = cluster_accuracy(y_true, y_pred)
ari = adjusted_rand_score(y_true.values.flatten(), y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"ARI: {ari:.4f}")
```
